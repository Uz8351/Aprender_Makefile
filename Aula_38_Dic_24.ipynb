```python
import math
import sqlite3

# Crear o conectar a la base de datos SQLite
conn = sqlite3.connect("operaciones.db")
cursor = conn.cursor()

# Crear la tabla para almacenar las operaciones
cursor.execute('''CREATE TABLE IF NOT EXISTS operaciones (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    operacion TEXT NOT NULL,
    num1 REAL,
    num2 REAL,
    resultado REAL
)''')
conn.commit()

def menu():
    """Muestra el menú de operaciones al usuario."""
    print("\n--- Tareas que hago ---")
    print("1. Sumo")
    print("2. Resto")
    print("3. Multiplico")
    print("4. Divido")
    print("5. Hago raíz cuadrada")
    print("6. Salir")

def registrar_operacion(operacion, num1, num2, resultado):
    """Guarda la operación en la base de datos."""
    cursor.execute("INSERT INTO operaciones (operacion, num1, num2, resultado) VALUES (?, ?, ?, ?)",
                   (operacion, num1, num2, resultado))
    conn.commit()

def sumar():
    """Realiza la suma de dos números ingresados por el usuario."""
    try:
        a = float(input("Introduce el primer número: "))
        b = float(input("Introduce el segundo número: "))
        resultado = a + b
        print(f"Resultado: {a} + {b} = {resultado}")
        registrar_operacion("Suma", a, b, resultado)
    except ValueError:
        print("Error, introduce un número válido")

def restar():
    """Realiza la resta de dos números ingresados por el usuario."""
    try:
        a = float(input("Introduce el primer número: "))
        b = float(input("Introduce el segundo número: "))
        resultado = a - b
        print(f"Resultado: {a} - {b} = {resultado}")
        registrar_operacion("Resta", a, b, resultado)
    except ValueError:
        print("Error, introduce un número válido")

def multiplicar():
    """Realiza la multiplicación de dos números ingresados por el usuario."""
    try:
        a = float(input("Introduce el primer número: "))
        b = float(input("Introduce el segundo número: "))
        resultado = a * b
        print(f"Resultado: {a} * {b} = {resultado}")
        registrar_operacion("Multiplicación", a, b, resultado)
    except ValueError:
        print("Error, introduce un número válido")

def dividir():
    """Realiza la división de dos números, manejando la división por cero."""
    try:
        a = float(input("Introduce el primer número: "))
        b = float(input("Introduce el segundo número: "))
        if b == 0:
            print("Error: No se puede dividir entre cero.")
        else:
            resultado = a / b
            print(f"Resultado: {a} / {b} = {resultado}")
            registrar_operacion("División", a, b, resultado)
    except ValueError:
        print("Error, introduce un número válido")

def raiz_cuadrada():
    """Calcula la raíz cuadrada de un número ingresado por el usuario."""
    try:
        a = float(input("Introduce un número: "))
        if a < 0:
            print("Error: No se puede calcular la raíz cuadrada de un número negativo.")
        else:
            resultado = math.sqrt(a)
            print(f"Resultado: √{a} = {resultado}")
            registrar_operacion("Raíz Cuadrada", a, None, resultado)
    except ValueError:
        print("Error, introduce un número válido")

def main():
    """Función principal que controla el flujo del programa."""
    while True:
        menu()
        opcion = input("Selecciona una opción: ")

        try:
            opcion = int(opcion)
            if opcion == 1:
                sumar()
            elif opcion == 2:
                restar()
            elif opcion == 3:
                multiplicar()
            elif opcion == 4:
                dividir()
            elif opcion == 5:
                raiz_cuadrada()
            elif opcion == 6:
                print("¡Hasta luego!")
                break
            else:
                print("Opción inválida. Por favor, selecciona un número del 1 al 6.")
        except ValueError:
            print("Entrada no válida. Por favor, introduce un número.")

if __name__ == "__main__":
    main()
    # Cerrar la conexión a la base de datos al terminar el programa
    conn.close()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Crear una matriz de ejemplo de 50 elementos
matriz = np.random.rand(10, 5)  # Genera una matriz de 10 filas y 5 columnas con valores aleatorios entre 0 y 1

# Configurar el tamaño de la figura y la matriz
fig, ax = plt.subplots(figsize=(8, 6))
ax.set_aspect('auto')

# Mostrar la matriz en una ventana emergente con regulación de scroll
ax.imshow(matriz, cmap='viridis')
ax.set_aspect('auto')
ax.autoscale(enable=True, tight=True)

# Agregar una barra de colores
cbar = ax.figure.colorbar(ax.get_images()[0])
cbar.ax.set_ylabel('Valores', rotation=90)

# Mostrar la figura
plt.show()
```

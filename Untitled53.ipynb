```python
import matplotlib.pyplot as plt

# Función para calcular el área de un triángulo
def calcular_area_triangulo(base, altura):
    """
    Calcula el área de un triángulo utilizando la fórmula (base * altura) / 2.
    :param base: La base del triángulo.
    :param altura: La altura del triángulo.
    :return: El área del triángulo.
    """
    area = (base * altura) / 2
    return area

# Función para calcular el perímetro de un rectángulo
def calcular_perimetro_rectangulo(ancho, largo):
    """
    Calcula el perímetro de un rectángulo utilizando la fórmula 2 * (ancho + largo).
    :param ancho: El ancho del rectángulo.
    :param largo: El largo del rectángulo.
    :return: El perímetro del rectángulo.
    """
    perimetro = 2 * (ancho + largo)
    return perimetro

# Función principal del programa
def main():
    """
    Función principal del programa. Solicita al usuario ingresar las dimensiones de un triángulo y un rectángulo,
    y muestra el área del triángulo, el perímetro del rectángulo y un gráfico de barras.
    """
    # Pedir al usuario que ingrese la base y la altura del triángulo
    base_triangulo = float(input("Ingrese la base del triángulo: "))
    altura_triangulo = float(input("Ingrese la altura del triángulo: "))

    # Calcular el área del triángulo utilizando la función calcular_area_triangulo
    area_triangulo = calcular_area_triangulo(base_triangulo, altura_triangulo)
    print("El área del triángulo es:", area_triangulo)

    # Pedir al usuario que ingrese el ancho y el largo del rectángulo
    ancho_rectangulo = float(input("Ingrese el ancho del rectángulo: ").replace(',', '.'))
    largo_rectangulo = float(input("Ingrese el largo del rectángulo: ").replace(',', '.'))

    # Calcular el perímetro del rectángulo utilizando la función calcular_perimetro_rectangulo
    perimetro_rectangulo = calcular_perimetro_rectangulo(ancho_rectangulo, largo_rectangulo)
    print("El perímetro del rectángulo es:", perimetro_rectangulo)

    # Crear el gráfico de barras
    fig, ax = plt.subplots()
    medidas = ['Área del triángulo', 'Perímetro del rectángulo']
    valores = [area_triangulo, perimetro_rectangulo]
    ax.bar(medidas, valores)

    # Añadir etiquetas a las barras
    for i, v in enumerate(valores):
        ax.text(i, v + 1, str(v), ha='center', va='bottom')

    # Configurar el título y los ejes del gráfico
    ax.set_title('Comparación de medidas')
    ax.set_xlabel('Medidas')
    ax.set_ylabel('Valores')

    # Mostrar el gráfico
    plt.show()

# Llamar a la función principal del programa
main()
```
